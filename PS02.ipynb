2) Write a class called SawtoothChirp that extends Chirp and overrides evaluate to generate a sawtooth waveform with frequency that increases (or decreases) linearly.

Hint: combine the evaluate functions from Chirp and SawtoothSignal.

Plot the spectrogram of this signal. The effect of aliasing should be visible, and if you listen carefully, you can hear it.

```python
class SawtoothChirp(Chirp):
    # Changing the actual evaluate function would be pointless since it actually uses _evaluate in the end
    
    def _evaluate(self, ts, freqs):
        '''
        ts: float array of times
        freqs: float array of frequencies during each interval
        '''
        # Copied from Chirp
        dts = np.diff(ts)
        dps = PI2 * freqs * dts
        phases = np.cumsum(dps)
        phases = np.insert(phases, 0, 0)
        
        frac, _ = np.modf(phases) # Takes the fractional part to use for the sawtoothing
        ys = normalize(unbias(frac), self.amp)
        return ys

sawtooth_chirp = SawtoothChirp(start=220, end=440, amp=1)
sawtooth_chirp_wave = sawtooth_chirp.make_wave(duration=2, start=0, framerate=44100)
sawtooth_chirp_wave.write('Sawtooth_Chirp_44k.wav')
sawtooth_chirp_spectrogram = sawtooth_chirp_wave.make_spectrogram(seg_length=512)
sawtooth_chirp_spectrogram.plot()

sawtooth_chirp_wave = sawtooth_chirp.make_wave(duration=2, start=0, framerate=22050)
sawtooth_chirp_wave.write('Sawtooth_Chirp_22k.wav')

sawtooth_chirp_wave = sawtooth_chirp.make_wave(duration=2, start=0, framerate=11025)
sawtooth_chirp_wave.write('Sawtooth_Chirp_11k.wav')

plt.xlabel("Frequency (Hz)")
plt.ylabel("Time (s)")
plt.title("Spectrogram of Sawtooth Chirp")
plt.show()
```

3)  In musical terminology, a “glissando” is a note that slides from one pitch to another, so it is similar to a chirp. A trombone player can play a glissando by extending the trombone slide while blowing continuously. As the slide extends, the total length of the tube gets longer, and the resulting pitch is inversely proportional to length.

Assuming that the player moves the slide at a constant speed, how does frequency vary with time? Is a trombone glissando more like a linear or exponential chirp?

Write a function that simulates a trombone glissando from C3 up to F3 and back down to C3. C3 is 262 Hz; F3 is 349 Hz.

```python
# Assume that the slide itself is a linear motion, 
# so the lengths follow something of a Triangular function
# as the Trombonist slides in then back out
# This order is chosen because the note goes low-high-low which corresponds to long-short-long

num = 1024
duration = 1.0 # Duration in seconds
F3 = 349
C3 = 262
base = 1.0 / F3
change = 1.0 / C3 - 1.0 / F3

# Creates an array of length num * 2 with "lengths" that range from 1/C3 to 1/F3
# then back to 1/C3 with two line segments by leveraging absolute values
trombone_lengths = [abs(1.0- 1.0 * x / num) * change + base for x in range(num * 2)]

times = [duration * x / num / 2 for x in range(num * 2)] # X Label for easier reading

pitches = [1.0 / length for length in trombone_lengths] # Generates the pitches which are "inversely proportional" to length
plt.plot(times, pitches)
plt.xlabel("Time (s)")
plt.ylabel("Pitch (Hz)")
plt.title('Glissando shortrange')
plt.show()
```

```python
# Same as the above, but with a more extreme pitch variance to determine whether it's linear or exponential
num = 1024
duration = 1.0 # Duration in seconds
F4 = F3 * 2 # One octave up
C2 = C3 / 2 # One octave down
base = 1.0 / F4
change = 1.0 / C2 - 1.0 / F4

# Creates an array of length num * 2 with "lengths" that range from 1/C3 to 1/F3
# then back to 1/C3 with two line segments by leveraging absolute values
trombone_lengths = [abs(1.0- 1.0 * x / num) * change + base for x in range(num * 2)]

times = [duration * x / num / 2 for x in range(num * 2)] # X Label for easier reading

pitches = [1.0 / length for length in trombone_lengths] # Generates the pitches which are "inversely proportional" to length
plt.plot(times, pitches)
plt.xlabel("Time (s)")
plt.ylabel("Pitch (Hz)")
plt.title('Glissando Longrange')
plt.show()
```

In Glissando Shortrange, there could be dispute whether that frequency change is linear or exponential in quality.  There is, without a doubt, a slight curve, but to call it exponential because of that would be missing many of its qualities.  However, looking at the second graph, Glissando Longrange, reveals that the behavior is more exponential in nature over more extreme changes.  However, two entire octaves are spanned during this time to go from C2 to F4.  The inversely proportional nature of the tones suggest that the quality is closer to an exponential chirp rather than a linear one.

6) In a white noise signal, the mixture of frequencies changes over time. In the long run, we expect the power at all frequencies to be equal, but in any sample, the power at each frequency is random.

To estimate the long-term average power at each frequency, we can break a long signal into segments, compute the power spectral density for each segment, and then compute the average across the segments. 

You can read more about this algorithm at http://en.wikipedia.org/wiki/Bartlett's_method.

Apply Bartlett's method to your natural noise source and see if it gives a clearer picture of the relationship between frequency and power.

Bartlett’s method consists of the following steps:

1. The original N point data segment is split up into K (non-overlapping) data segments, each of length M

2. For each segment, compute the periodogram by computing the discrete Fourier transform (DFT version which does not divide by M), then computing the squared magnitude of the result and dividing this by M.

3. Average the result of the periodograms above for the K data segments.
    The averaging reduces the variance, compared to the original N point data segment.

The end result is an array of power measurements vs. frequency "bin".

```python
fire_sound = read_wave("fireplace.wav")
units = 2500000
length = 100
fire_sound.truncate(units) # Trims to 2.5e6 units
# fire_segments = [0] * length
seg_length = 2500000 / length # 2.5e4 units per segment

final_spectrum = Spectrum([1], 1)

for i in range(length):
    segment = fire_sound.copy()
    segment.shift(-1 * seg_length * i)
    segment.truncate(seg_length)
    spectrum = segment.make_spectrum()
    if(i == 0):
        final_spectrum = spectrum.power
    else:
        final_spectrum = final_spectrum + spectrum.power

for i in range(length):
    final_spectrum[i] = final_spectrum[i] / length
        
plt.yscale('log')
plt.ylim([1e14, 1e18])
plt.xlim([0, 12500])
plt.title("Bartlett's Method on the noise of a Fireplace")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Average Power")
plt.plot(final_spectrum)
plt.show()
```

Computing the average power density of the fireplace sound shows that the sound is definitely not white noise because the average variance in the different frequencies is fairly huge.  There are two magnitudes of difference in the intensities of different frequencies, which is far from what is considered white noise.
